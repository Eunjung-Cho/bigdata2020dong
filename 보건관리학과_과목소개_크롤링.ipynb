```python
import urllib.request
from bs4 import BeautifulSoup
import re
import numpy as np
```

```python
print('먼저 과목 설명이 있는 페이지 각각의 링크를 수집하겠습니다')
url = 'https://health.dongduk.ac.kr/myboard/sub1_4'

soup = BeautifulSoup(urllib.request.urlopen(url).read(), 'html.parser')
stat_links = []
for link in soup.findAll("a"):
    if 'href' in link.attrs: 
        link = str(link.attrs['href'])
        if link[:5] == '/shop':
            stat_links.append(link)
stat_links = set(stat_links)
stat_links = list(stat_links)
print(stat_links)
        
```

```python
if __name__  == "__main__":
    contents = []
    for stat in stat_links:
        url = 'https://health.dongduk.ac.kr/'+stat
        #print('+------------크롤링중-----------+')
        #print('url:',url)
        soup = BeautifulSoup(urllib.request.urlopen(url).read(), 'html.parser')
        res = soup.find_all('table', 'res-bd-view')
        for n in res:
            content = ''
            con = n.get_text()
            con.strip()
            #con.replace("\n", " ")
            con = re.sub('[\n]+', " ",con)
            content = content+con
            
            contents.append(content)
            
            
    print(contents)
```

```python
import pandas as pd
df = pd.DataFrame(contents)
```

```python
df.head()
```

```python
df['학수번호'] = np.nan
df['학과'] = '보건관리학과'
```

```python
df.columns = ['과목소개','학수번호','학과']
```

```python
df = df[['학과','학수번호','과목소개']]
```

```python
df.loc[1,'과목소개'][19:27]
```

```python
for r in range(df.shape[0]):
    df.loc[r,'학수번호'] = df.loc[r,'과목소개'][19:27]
print(df)
```

```python
for r in range(df.shape[0]):
    df.loc[r,'과목소개'] = re.sub('기본 폼 구분 전공선택 학수번호',"",df.loc[r,'과목소개'])
    df.loc[r,'과목소개'] = re.sub('기본 폼 구분 전공필수 학수번호',"",df.loc[r,'과목소개'])
    df.loc[r,'과목소개'] = re.sub('학습목표',"",df.loc[r,'과목소개'])
    df.loc[r,'과목소개'] = re.sub('강의개요',"",df.loc[r,'과목소개'])
    df.loc[r,'과목소개'] = re.sub('교과목해설',"",df.loc[r,'과목소개'])
    df.loc[r,'과목소개'] = re.sub('평가방법',"",df.loc[r,'과목소개'])
    df.loc[r,'과목소개'] = re.sub('경영K00',"",df.loc[r,'과목소개'])
    df.loc[r,'과목소개'] = re.sub('경영B00',"",df.loc[r,'과목소개'])
    df.loc[r,'과목소개'] = re.sub('과목명',"",df.loc[r,'과목소개'])
    df.loc[r,'과목소개'] = re.sub(df.loc[r,'학수번호'],"",df.loc[r,'과목소개'])
    pattern = '[0-9]' # 숫자제거
    df.loc[r,'과목소개'] = re.sub(pattern=pattern, repl='', string=df.loc[r,'과목소개'] )
    pattern = '[^\w\s]'
    df.loc[r,'과목소개'] = re.sub(pattern=pattern, repl='', string=df.loc[r,'과목소개'])
    pattern = '[ㆍ]'
    df.loc[r,'과목소개'] = re.sub(pattern=pattern, repl='', string=df.loc[r,'과목소개'])
    df.loc[r,'과목소개'] = df.loc[r,'과목소개'][:-52]
    print(r,df.loc[r,'과목소개'])
print(df.head(3))
```
